```
from scipy import random
import numpy as np
import matplotlib.pyplot as plt
n=100                # no of random points
pi=[]   
circle_point=[]
square_point=[]           
for i in range(10000):
    m=0
    for i in range(n):
        x=random.uniform(0,1)
        y=random.uniform(0,1)
        d=np.sqrt(x**2+y**2)
        square_point.append(np.c_[x,y])
        if d<1:
            m=m+1
            circle_point.append(np.c_[x,y])
    pi.append(4*m/n)
plt.hist(pi,bins=10,ec='black')
print(np.sum(pi)/10000)
```

```
circle_point=np.array(circle_point).reshape([-1,2])
square_point=np.array(square_point).reshape([-1,2])
```

```
plt.scatter(circle_point[:1000,0],circle_point[:1000,1],color='red')
plt.scatter(square_point[:1000,0],square_point[:1000,1],color='blue')
```
